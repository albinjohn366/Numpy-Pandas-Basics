# NUMPY RANDOM
The following functions can be used to play around with numpy arrays and data. It showcases how random method in numpy is useful for creating pseudo random values and arrays. It also features other applications of random.

```
import numpy as np
```

```
# To randomly create n-dimensional array
a = np.random.choice([3, 5, 7, 9], size=(2, 3))
print(a)
```

```
# Random choice with probability
a = np.random.choice([2, 5, 12, 4], p=[0.1, 0.4, 0.3, 0.2], size=(100))
print(a)
```

```
# Shuffle and permutation
a = np.array([1, 2, 4, 5, 6])
np.random.shuffle(a)
print(a)

a = np.array([1, 2, 4, 5, 6])
a = np.random.permutation(a)
print(a)
```

```
# Seaborn to visualize
import seaborn as sns
import matplotlib.pyplot as plt
a = [12, 34, 2, 56, 21]
sns.distplot(a)
plt.show()

import seaborn as sns
import matplotlib.pyplot as plt
a = [12, 34, 2, 56, 21]
sns.distplot(a, hist=False)
plt.show()
```

```
# Normal distribution
a = np.random.normal(size=(4, 3), loc=4, scale=0.5)
print(a)
sns.distplot(a)
plt.show()
```

```
# Binomial distribution
a = np.random.binomial(size=10, p=0.5, n=10)
print(a)
sns.distplot(a)
plt.show()
```

```
# Poisson Distribution
a = np.random.poisson(size=100, lam=2)
print(a)
sns.distplot(a)
plt.show()
```

```
# Uniform Distribution
a = np.random.uniform(size=(100))
print(a)
sns.distplot(a)
plt.show()
```

```
a = np.random.logistic(scale=0.5, loc=40, size=(20, 4))
print(a)
sns.distplot(a)
plt.show()
```

```
# Multinomial Distribution
a = np.random.multinomial(n=5, pvals=[1/5, 1/5, 1/5, 1/5, 1/5])
print(a)
sns.distplot(a)
plt.plot()
```

```
# Exponential distribution
a = np.random.exponential(scale=3, size=(100))
print(a)
sns.distplot(a)
plt.show()
```

```
# Chi square
a = np.random.chisquare(size=30, df=2)
print(a)
sns.distplot(a)
plt.show()
```

```
a = np.random.rayleigh(scale=4, size=30)
print(a)
sns.distplot(a)
plt.show()
```

```
a = np.random.zipf(a=2, size=8)
print(a)
sns.distplot(a)
plt.plot()
```
